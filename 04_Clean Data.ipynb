```python
import pandas as pd
```

```python
df_amazon = pd.read_csv('amazon.csv')
df_amazon_jabon = pd.read_csv('amazon_jabon.csv')
df_purenature = pd.read_csv('purenature.csv')
```

```python
df = pd.concat([df_amazon, df_amazon_jabon, df_purenature])
```

```python
df.product_price = df.product_price.replace({'€': '', ',': '.', ' ': '', 'desde\n': ''}, regex=True).astype(float)
df.link = df['link'].apply(lambda x: 'https://amazon.com' + x if x.startswith('/') else x)
```

```python
columnas_nombre = ['Categoría', 'Nombre', 'Tienda', 'Precio Normalizado', 'Precio', 'Cantidad', 'Imagen', 'URL', 'ID']
df.columns = columnas_nombre
```

```python
columns_order = ['Imagen', 'Nombre', 'Categoría', 'Tienda', 'Precio', 'URL', 'ID', 'Precio Normalizado']

df = df[columns_order].copy()

df = df.set_index('Imagen')
```

```python
df.to_csv('data_clean.csv', index=True)
```
